```python
import numpy as np
from scipy.stats import norm
from scipy.integrate import quad

# Define the two random variables
X = norm(0, 1)
Y = norm(0, 2)

# Define the joint PDF function
def joint_pdf(x, y):
    return X.pdf(x) * Y.pdf(y)

# Set the integration limits for X and Y
x_lower = -3
x_upper = 3
y_lower = -6
y_upper = 6

# Calculate the marginal probability of X
def marginal_pdf_x(x):
    return quad(lambda y: joint_pdf(x, y), y_lower, y_upper)[0]

# Calculate the marginal probability of Y
def marginal_pdf_y(y):
    return quad(lambda x: joint_pdf(x, y), x_lower, x_upper)[0]

# Calculate the conditional probability of Y given X
def conditional_pdf_y_given_x(y, x):
    return joint_pdf(x, y) / marginal_pdf_x(x)

# Calculate the conditional probability of X given Y
def conditional_pdf_x_given_y(x, y):
    return joint_pdf(x, y) / marginal_pdf_y(y)

# Perform the integration to calculate marginal PDFs
marginal_pdf_X, _ = quad(marginal_pdf_x, x_lower, x_upper)
marginal_pdf_Y, _ = quad(marginal_pdf_y, y_lower, y_upper)

print("Marginal PDF of X:", marginal_pdf_X)
print("Marginal PDF of Y:", marginal_pdf_Y)

# Choose some specific values for X and Y for conditional probability calculation
x_value = 0
y_value = 1

# Calculate conditional probabilities
cond_prob_Y_given_X = conditional_pdf_y_given_x(y_value, x_value)
cond_prob_X_given_Y = conditional_pdf_x_given_y(x_value, y_value)

print("Conditional Probability of Y given X:", cond_prob_Y_given_X)
print("Conditional Probability of X given Y:", cond_prob_X_given_Y)
```

This code performs calculations related to joint, marginal, and conditional probabilities of two random variables X and Y, based on their probability density functions (PDFs).

Let's break down the code step by step:

1. **Importing Libraries**: 
   - `numpy` as `np`: Used for numerical computations.
   - `norm` from `scipy.stats`: Represents a normal distribution.
   - `quad` from `scipy.integrate`: Performs numerical integration.

2. **Defining Random Variables**:
   - `X = norm(0, 1)`: Defines a normal distribution with mean 0 and standard deviation 1.
   - `Y = norm(0, 2)`: Defines a normal distribution with mean 0 and standard deviation 2.

3. **Defining the Joint PDF Function**:
   - `def joint_pdf(x, y):`: Defines a function to compute the joint probability density function (PDF) of X and Y.
   - `return X.pdf(x) * Y.pdf(y)`: Computes the joint PDF by multiplying the individual PDFs of X and Y.

4. **Setting Integration Limits**:
   - `x_lower = -3`, `x_upper = 3`: Define the integration limits for variable X.
   - `y_lower = -6`, `y_upper = 6`: Define the integration limits for variable Y.

5. **Calculating Marginal PDFs**:
   - `def marginal_pdf_x(x):`: Defines a function to calculate the marginal probability density function (PDF) of X.
   - `def marginal_pdf_y(y):`: Defines a function to calculate the marginal PDF of Y.
   - `quad(lambda y: joint_pdf(x, y), y_lower, y_upper)`: Performs numerical integration over Y to obtain the marginal PDF of X.
   - `quad(lambda x: joint_pdf(x, y), x_lower, x_upper)`: Performs numerical integration over X to obtain the marginal PDF of Y.

6. **Calculating Conditional Probabilities**:
   - `def conditional_pdf_y_given_x(y, x):`: Defines a function to calculate the conditional probability density function of Y given X.
   - `def conditional_pdf_x_given_y(x, y):`: Defines a function to calculate the conditional PDF of X given Y.
   - These conditional probabilities are computed by dividing the joint PDF by the respective marginal PDF.

7. **Performing Integrations to Calculate Marginal PDFs**:
   - `marginal_pdf_X, _ = quad(marginal_pdf_x, x_lower, x_upper)`: Performs numerical integration to obtain the marginal PDF of X.
   - `marginal_pdf_Y, _ = quad(marginal_pdf_y, y_lower, y_upper)`: Performs numerical integration to obtain the marginal PDF of Y.

8. **Printing Results**:
   - Prints the calculated marginal PDFs and conditional probabilities for given values of X and Y.

This code provides a framework for calculating various probability-related quantities based on the joint PDF of two random variables X and Y.
