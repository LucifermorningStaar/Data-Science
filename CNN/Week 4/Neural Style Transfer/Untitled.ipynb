```python
import tensorflow as tf

# Sample tensor with shape [batch_size, height, width, channels]
a_C = tf.random.normal(shape=(2, 32, 32, 3))  # Replace with your actual tensor

# Transpose with the specified permutation
transposed_a_C = tf.transpose(a_C, perm=[0, 3, 1, 2])

# Print the original and transposed shapes
print("Original shape:", a_C.shape)
print("Transposed shape:", transposed_a_C.shape)
```

```python
import math
from turtle import *

def hearta(k):
    return 15*math.sin(k)**3

def heartb(k):
    return 12*math.cos(k)-5*\
           math.cos(2*k)-2*\
           math.cos(3*k)-\
           math.cos(4*k)
speed(1000)
bgcolor("red")
for i in range (6000):
    goto(hearta(i)*20,heartb(i)*20)
    for j in range (5):
        #color("f73487")
        goto(0,0)
        
done()    
```
